```python
#데이터 불러오기
dataset_path = Path("dataset")
item = "슬리브리스"
item_dir = Path(os.path.join(dataset_path,item))
og_dir =  Path(os.path.join(dataset_path,item + ' copy'))
def get_file_paths(directory):
    files_path = []
    if os.path.exists(directory):
        files_path = [str(file) for file in directory.rglob('*')]
        files_path = sorted(files_path, key=lambda x: int(re.search(r'p(\d+)', x).group(1)) if re.search(r'p(\d+)', x) else float('inf'))
    else:
        print(f"directory not found")
        return
    return files_path
item_paths = get_file_paths(item_dir)
og_paths = get_file_paths(og_dir)
```

```python
# 파일 불러오기 및 저장, DataFrame 관련 헬퍼 함수

def get_file(path):
    """
    주어진 경로에서 CSV 파일을 읽어 Pandas DataFrame으로 반환하고,
    '제품_색깔', '제품_사이즈', '제품_재질' 열을 추가로 생성합니다.
    
    Args:
        path (str): 파일 경로
    Returns:
        pd.DataFrame: 파일에서 로드된 데이터와 추가된 컬럼
    """
    df = pd.read_csv(path, on_bad_lines="skip")  # 잘못된 행 건너뛰며 파일 로드
    df['제품_색깔'] = None  # '제품_색깔' 열 추가 (초기값: None)
    df['제품_사이즈'] = None  # '제품_사이즈' 열 추가 (초기값: None)
    df['제품_재질'] = None  # '제품_재질' 열 추가 (초기값: None)
    return df


def save_file(df, path):
    """
    DataFrame을 주어진 경로에 CSV 파일로 저장합니다.

    Args:
        df (pd.DataFrame): 저장할 데이터프레임
        path (str): 저장 경로
    """
    df.to_csv(path, index=False)  # 인덱스 없이 저장


def get_option(option):
    """
    'og' 또는 'replace' 옵션에 따라 파일 경로를 반환합니다.

    Args:
        option (str): 파일 경로 옵션 ('og' 또는 'replace')
    Returns:
        list: 선택된 옵션의 파일 경로 목록
    """
    if option == "og":
        return og_paths  # 원본 파일 경로 반환
    elif option == "replace":
        return item_paths  # 대체 파일 경로 반환
    else:
        return f"no option exists"  # 잘못된 옵션에 대한 알림


# DataFrame 출력 옵션 토글 함수

def toggle_display(show_all=True):
    """
    DataFrame 출력 설정을 조정합니다. 
    
    Args:
        show_all (bool): True일 경우 모든 행과 열을 표시, False일 경우 기본 설정으로 복원
    """
    if show_all:
        pd.set_option('display.max_rows', None)  # 모든 행 표시
        pd.set_option('display.max_columns', None)  # 모든 열 표시
        pd.set_option('display.max_colwidth', None)  # 컬럼 내용 전체 표시
        pd.set_option('display.expand_frame_repr', False)  # 데이터프레임 줄바꿈 방지
    else:
        pd.reset_option('display.max_rows')  # 행 출력 설정 기본값 복원
        pd.reset_option('display.max_columns')  # 열 출력 설정 기본값 복원
        pd.reset_option('display.max_colwidth')  # 컬럼 내용 출력 기본값 복원
        pd.reset_option('display.expand_frame_repr')  # 데이터프레임 줄바꿈 설정 복원


# 코드 실행 예제

if item_paths:  # 'item_paths'가 비어있지 않은 경우 실행
    sample_df = get_file(item_paths[0])  # 첫 번째 파일 로드 및 새로운 컬럼 추가
    # print(sample_df.head())  # 새로운 컬럼 포함 데이터프레임 첫 5행 출력 (필요 시 활성화)
```

```python
# 옵션 데이터 분류 함수
def categorize_options(df):
    """
    옵션 데이터를 '제품_색깔', '제품_사이즈', '제품_재질'로 분류하여 각각의 열에 저장합니다.

    Args:
        df (pd.DataFrame): 분류할 DataFrame
    Returns:
        pd.DataFrame: 분류된 결과가 반영된 DataFrame
    """
    # 정규식 패턴 정의
    color_pattern = r"(아이보리|블랙|크림|그레이|코발트블루|소라|핑크|백멜란지|베이지|브라운|모카)"
    size_pattern = r"(one size|One Color size|L)"
    material_pattern = r"(기모|롱ver|롱롱ver|벨로아|앙고라기모|long ver|Long ver)"

    # 색깔, 사이즈, 재질 열 생성
    df['제품_색깔'] = df['옵션1'].apply(lambda x: re.search(color_pattern, x).group(1) if re.search(color_pattern, x) else None)
    df['제품_사이즈'] = df['옵션2'].apply(lambda x: re.search(size_pattern, x).group(1) if re.search(size_pattern, x) else None)
    df['제품_재질'] = df['옵션1'].apply(lambda x: re.search(material_pattern, x).group(1) if re.search(material_pattern, x) else None)

    return df
df
```

```python
# 수동적으로 해야할 것들
# Check for 키, 몸무게, 사이즈 rows where the unit is missing or invalid

unit_dict = {'키': 'cm', '몸무게': 'kg', '사이즈': '상의'}

def check_wrong_data(df):
    invalid_rows_list = []
    
    for col in unit_dict.keys():
        expected_unit = unit_dict[col]
        invalid_rows = df[df[col].apply(lambda x: pd.notna(x) and isinstance(x, str) and expected_unit not in x)]
        invalid_rows_list.append(invalid_rows)
    
    # Combine all invalid rows and drop duplicates
    consolidated_invalid_rows = pd.concat(invalid_rows_list).drop_duplicates()
    
    return consolidated_invalid_rows

def check_wrong_items(option):
    
    paths = get_option(option)
    
    combined_df = pd.DataFrame()
    
    for path in paths:
        if path.endswith('.txt'):
            continue
        df = get_file(path)
        wrong_df = check_wrong_data(df)
        combined_df = pd.concat([combined_df, wrong_df], ignore_index=True)
    return combined_df



# Check Categorical Data
def check_cat_values(option):
    
    # Turn on 표 전체보기 
    toggle_display(True)
    
    check_cols = ["리뷰어 등급","옵션1","옵션2","상품 사이즈","퀄리티","색감"]
    paths = get_option(option)

    summary = pd.DataFrame() # Create empty dataframe
    file_summary = {}
    idx = 1
    
    for path in paths:
        
        if path.endswith(".txt"): #.txt 파일은 무시 
            continue
        
        df = get_file(path) 
        
        for col in check_cols:
            value_counts = df[col].value_counts()
            for value, count in value_counts.items():
                file_summary[(idx, col, value)] = count
                
        idx = idx + 1
        value_ct_df = pd.DataFrame.from_dict(file_summary, orient="index", columns=["Count"])
        index = pd.MultiIndex.from_tuples(value_ct_df.index, names=["Product", "Column", "Value"])
        summary_df = pd.DataFrame(value_ct_df,index=index).reset_index()
        grouped_df = pd.DataFrame(summary_df.groupby(["Product","Column"]).agg(list))
        
    # Combine summaries into a single DataFrame
    summary = pd.concat([summary,grouped_df], axis = 0)
    summary = summary.sort_values(by=["Product", "Column"])
    return summary
   
        
       
        

```

```python
# 위 두개를 돌려보고 난 뒤
# 긴소매티셔츠 기준 , unit 기준으로 대체해도 괜찮다고 판단
def move_data(df):
    for index, row in df.iterrows():
        for col in unit_dict.keys():
            expected_unit = unit_dict[col]
            if pd.notna(row[col]) and expected_unit not in str(row[col]):
                # Find the correct column for this value
                for target_col, target_unit in unit_dict.items():
                    if target_unit in str(row[col]):
                        # Move the value to the correct column
                        df.at[index, target_col] = row[col]
                        df.at[index, col] = None  # Clear the original value
                        break
    return df

def move_all_data(option):
    if option == "og":
        paths = og_paths
    elif option == "replace":
        paths = item_paths
    else:
        return (f"no option exists")
    
    for path in item_paths:
        if path.endswith('.txt'):
            continue
        df = get_file(path)
        updated_df = move_data(df) # 여기서 실제 inplace
        save_file(updated_df, path) # 그리고 저장
    
    
```
